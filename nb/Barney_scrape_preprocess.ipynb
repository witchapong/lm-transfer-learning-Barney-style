```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import re
import pandas as pd
import os
```

```python
# target url for scraping
base_url = "http://transcripts.foreverdreaming.org/viewforum.php?f=177"
season_list = ['01','02','03','04','05','06','07','08','09']
anchor_txt = []
href = []

for i in range(9):
    
    if i!=0:
        url = base_url + f'&start={i*25}'
    else:
        url = base_url
        
    # create a new chrome
    driver = webdriver.Chrome()
    driver.implicitly_wait(30)

    # get the html data from the url
    driver.get(url)

    # parse the html data with BeautifulSoup to get the DOM object
    soup_level1=BeautifulSoup(driver.page_source, 'lxml')

    #driver.close()

    # the return item is as a list
    soup_level2=soup_level1.find_all('td',class_='topic-titles row2')
    
    for td in soup_level2:
        if td.a.text[:2] in season_list:
            anchor_txt.append(td.a.text)
            href.append(td.a['href'])
```

```python
base_url = "http://transcripts.foreverdreaming.org"
barney_conv=[]
driver = webdriver.Chrome()

for suburl in href:
    driver.implicitly_wait(30)
    driver.get(base_url+suburl[1:])
    soup_level1=BeautifulSoup(driver.page_source, 'lxml')

    for p in soup_level1.findAll('p'):
        if p.attrs == {}:
            if 'Barney: ' in p.text:
                barney_conv.append(p.text.split('Barney: ')[1])
```

```python
barney_df = pd.DataFrame({'txt':barney_conv})
```
